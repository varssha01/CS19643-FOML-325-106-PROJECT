```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
df = pd.read_csv('Salary_Data.csv')
df.head()
```

```python
X = df[['YearsExperience']].values  # Assuming 'column_2' is the name of the second column
if 'Salary' in df.columns:  # Check if target column exists
    y = df['Salary'].values
else:
    print("Target column doesn't exist")

from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
plt.scatter(X, y, color="red")
plt.plot(X, lin_reg.predict(X))
plt.title("Linear Regression")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.show()
```

```python

from sklearn.preprocessing import PolynomialFeatures
poly_reg_2 = PolynomialFeatures(degree=3)
X_poly_2 = poly_reg_2.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly_2, y)
plt.scatter(X, y, color="red")
plt.plot(X, lin_reg_2.predict(X_poly_2))
plt.title("Poly Regression - Degree 3")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.show()
```

```python
import numpy
import matplotlib.pyplot as plt

x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

mymodel = numpy.poly1d(numpy.polyfit(x, y, 3))

myline = numpy.linspace(1, 22, 100)

plt.scatter(x, y)
plt.plot(myline, mymodel(myline),color="red")
plt.show()
```
